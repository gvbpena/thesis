```python
from pydub import AudioSegment
import os

def split_audio(input_file, output_dir, last_segment_number):
    # Load the audio file
    audio = AudioSegment.from_wav(input_file)
    
    # Calculate duration in milliseconds
    duration_ms = len(audio)
    
    # Split the audio into 30-second segments
    segment_length_ms = 30 * 1000  # 30 seconds in milliseconds
    num_segments = duration_ms // segment_length_ms
    
    # Check if the audio is at least 30 seconds long
    if num_segments == 0:
        print(f"Skipping {input_file}: Audio is less than 30 seconds long.")
        return
    
    # Get the title from the input file name
    title = os.path.splitext(os.path.basename(input_file))[0]
    
    # Start numbering the segments from the last recorded segment number
    segment_number = last_segment_number + 1
    
    # Iterate through the segments
    for i in range(num_segments):
        start_time = i * segment_length_ms
        end_time = start_time + segment_length_ms
        segment = audio[start_time:end_time]
        
        # Name the output file with title and segment number
        output_file = f"{output_dir}/metal.{segment_number:05} - {title}.wav"

        
        # Export the segment to a WAV file
        segment.export(output_file, format="wav")
        
        segment_number += 1
    
    return segment_number - 1

def split_audio_files_in_folder(input_folder, output_folder):
    # Create the output directory if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)
    
    # Initialize the last segment number
    last_segment_number = 99
    
    # Iterate through each WAV file in the input folder
    for filename in os.listdir(input_folder):
        if filename.endswith(".wav"):
            input_file = os.path.join(input_folder, filename)
            last_segment_number = split_audio(input_file, output_folder, last_segment_number)

# Input and output folder paths
input_folder = "original_music/metal"  # Replace with the path to your input folder
output_folder = "metal"  # Path to save the segmented audio files

# Split audio files in the input folder and save to the output folder
split_audio_files_in_folder(input_folder, output_folder)
```
